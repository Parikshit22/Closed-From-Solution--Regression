```python
import numpy as np
from sklearn.datasets import make_regression
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X,Y = make_regression(n_samples = 400,n_features=2,n_informative=1,noise = 1.8, random_state = 10)
print(X.shape)
Y = Y.reshape(-1,1)
X = (X-X.mean())/X.std()
print(Y.shape)
```

```python
fig = plt.figure(figsize = (10,10))
axes = fig.gca(projection = '3d')
axes.scatter(X[:,0],X[:,1],Y,cmap = 'rainbow')
plt.show()
```

```python
m = X.shape[0]
one = np.ones((m,1))
x_data = np.concatenate((X,one),axis = 1)
print(x_data.shape)
```

```python
def Predict(X,theta):
    return np.dot(X,theta)
def close_form(X,Y):
    Y = np.mat(Y)
    first_term = np.dot(X.T,X)
    second_term = np.dot(X.T,Y)
    first_term = np.linalg.pinv(first_term)
    theta = first_term*second_term
    return theta
theta = close_form(x_data,Y)
print(theta)
pred = Predict(x_data,theta)
```

```python
fig = plt.figure(figsize = (10,10))
axes = fig.gca(projection = '3d')
axes.scatter(X[:,0],X[:,1],pred,color = 'r')
axes.scatter(X[:,0],X[:,1],Y,alpha = 0.)
plt.show()
```
